Inheritance atau biasa disebut dengan pewarisan adalah konsep pemrograman dimana sebuah class dapat menurunkan property dan method yang dimilikinya kepada class lain. Class yang akan diturunkan bisa disebut sebagai class induk (parent class) Sedangkan class yang menerima penurunan bisa disebut sebagai class anak (child class) atau juga sub class .
